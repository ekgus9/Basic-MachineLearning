# 21. 질의 응답 Question Answering, QA

### 메모리 네트워크 Memory Network, MemN 를 이용한 QA

image.png

<그림> 메모리 네트워크 구조를 간단히 표현한 그림

그림의 상단을 보면 스토리 문장과 질문 문장, 두 개의 입력이 들어온다. 임베딩 된 두 문장은 내적을 통해 각 단어 간 유사도를 구하고, 이렇게 구해진 유사도는 소프트 맥스 함수를 지나 Embedding A로 임베딩된 스토리 문장에 더해진다. 

이를 다르게 설명하자면 스토리 문장을 Value와 Key라 하고 질문 문장을 Query라 하자. query는 key와 유사도를 구하고, 소프트 맥스 함수를 통해 값을 정규화하여 value에 더해 유사도 값을 반영한다. 이러한 연산은 **어텐션 메커니즘**의 의도를 갖고 있다. 

이처럼 스토리 문장 표현을 얻었다면 이 문장을 질문 문장을 임베딩한 질문 표현과 연결해준다. 그리고 이를 LSTM과 밀집층의 입력으로 사용하여 정답을 예측한다.
